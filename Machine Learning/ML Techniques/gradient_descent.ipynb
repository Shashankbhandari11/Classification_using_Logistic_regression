```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
X, y = make_regression(n_samples = 4, n_features= 1,n_informative = 1,n_targets= 1,noise = 80, random_state = 13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(X,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='blue')
```

```python
#To find the intercept s,
#Assuming the intercept starting value as b = 0
y_pred = (78.35 *X + 0).reshape(4)
```

```python
print(y_pred)
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='Brown',label = 'OLS')
plt.plot(X,y_pred,color='blue',label='b = 0')
plt.legend()
plt.show()
```

```python
m= 78.35
b= 0

loss_slope = -2 * np.sum(y- m*X.ravel()-b)
loss_slope
```

```python
lr = 0.1
step_size = loss_slope * lr 
step_size
```

```python
b = b - step_size
b
```

```python
y_pred1 = (78.35 * X + b).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='blue', label='OLS')
plt.plot(X,y_pred1,color='black',label='b = {}'.format(b))
plt.plot(X,y_pred,color='yellow',label='b = 0')
plt.legend()
plt.show()
```

```python
#Iteration 2
loss_slope = -2 * np.sum(y - m * X.ravel() - b) 
loss_slope
```

```python
step_size = loss_slope * lr
step_size
```

```python
b = b - step_size
b
```

```python
y_pred2 = ((78.35 * X) + b).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='black',label='OLS')
plt.plot(X,y_pred2,color='blue',label='b = {}'.format(b))
plt.plot(X,y_pred1,color='yellow',label='b = {}'.format(b))
plt.plot(X,y_pred,color='green',label='b = 0')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.datasets import make_moons

# Generate a non-linearly separable dataset
X, y = make_moons(n_samples=100, noise=0.1, random_state=42)

# Train SVM with linear kernel (Fails for non-linear data)
svm_linear = SVC(kernel='linear')
svm_linear.fit(X, y)

# Train SVM with RBF kernel (Uses Kernel Trick)
svm_rbf = SVC(kernel='rbf', gamma=1)
svm_rbf.fit(X, y)

# Plot the decision boundaries
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm')
plt.title("Non-Linearly Separable Data with SVM")
plt.show()
```
